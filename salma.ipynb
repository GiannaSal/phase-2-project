## 3 budget categories

```python
#calculating budget ranges
combined_movies_finances['budget'] = combined_movies_finances['budget'] \
    .replace('[\$,]', '', regex=True) \
    .astype(float) ##convert strin to float
min_budget = combined_movies_finances['budget'].min()
max_budget = combined_movies_finances['budget'].max()
budget_range = max_budget - min_budget
category_size = budget_range / 3
increment = 0.1

#establishing budget ranges
low_budget = [min_budget, round(min_budget + category_size, 1)]
medium_budget = [round(low_budget[1] + increment, 1), round(min_budget + 2*category_size + increment, 1)]
high_budget = [round(medium_budget[1] + increment, 1), max_budget]


print(low_budget)
print(medium_budget)
print(high_budget)
```

```python
combined_movies_finances['budget_category'] = pd.cut(
    combined_movies_finances['budget'],
    bins=6,
    labels=['Very Low', 'Low', 'Lower-Mid', 'Upper-Mid', 'High', 'Very High'],
    include_lowest=True
)

combined_movies_finances['budget_category'].value_counts()
```

## remove duplicates

```python
#convert lists to tuples
def lists_to_tuples(df):
    for col in df.columns:
        # Check if any row in the column is a list
        if df[col].apply(lambda x: isinstance(x, list)).any():
            df[col] = df[col].apply(lambda x: tuple(x) if isinstance(x, list) else x)
    return df

#convert lists to tuples
merged_movies = lists_to_tuples(merged_movies)
movie_info = lists_to_tuples(movie_info)

#drop duplicates
merged_movies = merged_movies.drop_duplicates()
movie_info = movie_info.drop_duplicates()
```

### count of genre prominence within each budget bin

```python
merged_movies['genres'].head(10)
merged_movies['genres'].apply(type).value_counts()
merged_movies = merged_movies[merged_movies['genres'].notna()]
```

```python
genre_counts = (
    merged_movies
    .groupby(['budget_category', 'genres'])
    .size()
    .reset_index(name='count')
)
```

```python
top_genres_per_bin = (
    genre_counts
    .groupby('budget_category', observed=True, group_keys=False)
    .apply(lambda x: x.nlargest(3, 'count'))  # top 3 per group by count
    .reset_index(drop=True)
)

print(top_genres_per_bin)
```

```python
merged_movies['domestic_gross_budget'] = pd.to_numeric(merged_movies['domestic_gross_budget'], errors='coerce')
merged_movies['worldwide_gross_budget'] = pd.to_numeric(merged_movies['worldwide_gross_budget'], errors='coerce')
```

```python
merged_movies['vote_average'].notna().value_counts().head()
```

```python
performance_summary = (
    merged_movies.groupby(['budget_category', 'genres'], observed=True)
              .agg(
                  avg_rating=('vote_average', 'mean'),
                  median_rating=('vote_average', 'median'),
                  movie_count=('title', 'count')
              )
              .reset_index()
)

top3_per_category = (
    performance_summary.groupby('budget_category', observed=True, group_keys=False)
                       .apply(lambda x: x.nlargest(3, 'movie_count'))  # or 'avg_rating' if you prefer
                       .reset_index(drop=True)
)

print(top3_per_category)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot top 3 genres per budget category by movie count
plt.figure(figsize=(10,6))
sns.barplot(
    data=top3_per_category,
    y='genres',  # genres on y-axis
    x='movie_count',  # size represents number of movies
    hue='budget_category',
    dodge=True
)
plt.title("Top 3 Genres per Budget Category by Movie Count")
plt.xlabel("Number of Movies")
plt.ylabel("Genre")
plt.show()
```
